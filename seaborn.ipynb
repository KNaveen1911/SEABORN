Seaborn:::is a data visualization library built on top of matplotlib and closely integrated with pandas data structures in Python.Visualization is the central part of Seaborn which helps in exploration and understanding of data.... Visualizing univariate and bivariate distribution


# distribution plots

```python
import seaborn as sns
import matplotlib.pyplot as plt

df=sns.load_dataset("tips")
print(df)
```

```python
print(df.corr())
```

```python
print(sns.heatmap(df.corr()))
```

####  jointplot::relation ship between 2 numeric values


```python
 sns.jointplot(x='tip',
    y='total_bill',
    data=df,kind='scatter',color='g',
    height=5, ratio=2,space=0.3) #kind=reg,scatter,hex
```

#### pairplot::also called scatterplot  i.e for 3D and 4D


```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='tip')
```

```python
print(df['smoker'].value_counts())
```

```python
df.info()
```

```python
sns.displot(df['tip'])
```

```python
#kde==kernal density estimation
sns.kdeplot(df['tip'])
```

```python
sns.histplot(df['tip'])
```

# categorical plots

```python
df=sns.load_dataset("tips")
```

```python
sns.countplot(x='day',data=df)  # x= sex,tip,day.etc....
plt.show()
```

```python
sns.barplot(x='day',y='tip',data=df)
plt.show()
```

```python
sns.boxplot(x='sex',y='tip',data=df)
```

```python
sns.boxplot(x='day',y='total_bill',data=df,palette='rainbow')
```

```python
sns.violinplot(x='day',y='total_bill',data=df)
```
